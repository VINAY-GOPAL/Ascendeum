```python
bidder_stat = df.groupby('bidder').agg(total_bids=('response_count', 'sum'),total_wins=('prebid_win_count', 'sum')).dropna()
bidder_stat['win_rate'] = (bidder_stat['total_wins'] / bidder_stat['total_bids'])*100
bidder_stat = bidder_stat.sort_values(by='win_rate', ascending=False)
bidder_stat
```
